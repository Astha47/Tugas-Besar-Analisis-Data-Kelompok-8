## Hubungan Penjualan Tiap Region

#### Penjualan Tiap Wilayah

```python
def SUM(tipe):
    Sales = sample[tipe]
    jumlah = 0
    for i in range(len(Sales)):
        jumlah = jumlah + Sales.iloc[i]
    return jumlah

NA = SUM('NA_Sales')
EU = SUM('EU_Sales')
JP = SUM('JP_Sales')
Other = SUM('Other_Sales')

salesTP = [['NA',NA],['EU',EU],['JP',JP],['Other',Other]]
sales = pd.DataFrame(salesTP, columns=['Region','Total Sales'])
sales.plot(kind="bar",x="Region",title="Sales")
```

### Intesifitas Produksi Publisher Spesifik

```python
pilihan_publisher =int(input("Masukkan pilihan publisher : "))
keyword = list_publisher['Publisher'].iloc[pilihan_publisher]

#memuat data publisher
hasil_search = data.loc[data['Publisher'] == keyword]
publish_publisher = hasil_search.value_counts('Year')

input_data_publisher = {}
input_data_publisher["Tahun"] = [0 for i in range(len(publish_publisher))]
input_data_publisher["Pertambahan"] = [0 for i in range(len(publish_publisher))]

for i in range(len(publish_publisher)):
    input_data_publisher["Tahun"][i] = publish_publisher.index[i]
for j in range(len(publish_publisher)):
    input_data_publisher["Pertambahan"][j] = publish_publisher.iloc[j]
data_publisher = pd.DataFrame(data=input_data_publisher)

data_publisher_sorted = data_publisher.sort_values('Tahun',ascending=True)

input_data = {}
input_data["Tahun"] = [0 for i in range(len(data_publisher_sorted))]
input_data["Jumlah"] = [0 for i in range(len(data_publisher_sorted))]

for i in range(len(data_publisher_sorted)):
    input_data["Tahun"][i] = data_publisher_sorted['Tahun'].iloc[i]
jumlah = 0

for j in range(len(data_publisher_sorted)):
    jumlah += data_publisher_sorted['Pertambahan'].iloc[j]
    input_data["Jumlah"][j] = jumlah
JumlahTiapTahun = pd.DataFrame(data=input_data)

title1 = "Produksi Game " + keyword
title2 = "Jumlah Game "+ keyword
data_publisher.sort_values('Tahun').plot(kind="line",x="Tahun",y="Pertambahan",title=title1)
JumlahTiapTahun.plot(kind="line",x="Tahun",y="Jumlah", title=title2)
```
