# Visualisasi Data

## Hubungan Penjualan Tiap Region

#### Penjualan Tiap Wilayah

```python
def SUM(tipe):
    Sales = sample[tipe]
    jumlah = 0
    for i in range(len(Sales)):
        jumlah = jumlah + Sales.iloc[i]
    return jumlah

NA = SUM('NA_Sales')
EU = SUM('EU_Sales')
JP = SUM('JP_Sales')
Other = SUM('Other_Sales')

salesTP = [['NA',NA],['EU',EU],['JP',JP],['Other',Other]]
sales = pd.DataFrame(salesTP, columns=['Region','Total Sales'])
sales.plot(kind="bar",x="Region",title="Sales")
```

#### Visualisasi Data Pertambahan Produksi Game Tiap Tahun

```python
#umlah game setiap tahun


DFJumlah = data.value_counts("Year")

input_data = {}
input_data["Tahun"] = [0 for i in range(len(DFJumlah))]
input_data["Pertambahan"] = [0 for i in range(len(DFJumlah))]

for i in range(len(DFJumlah)):
    input_data["Tahun"][i] = DFJumlah.index[i]
for j in range(len(DFJumlah)):
    input_data["Pertambahan"][j] = DFJumlah.iloc[j]
NewDataFrame = pd.DataFrame(data=input_data)

NewJumlahDataFrame = pd.DataFrame(data=input_data)

DataPertambahan = NewDataFrame.sort_values('Tahun',ascending=True)

input_data = {}
input_data["Tahun"] = [0 for i in range(len(DataPertambahan))]
input_data["Jumlah"] = [0 for i in range(len(DataPertambahan))]

for i in range(len(DataPertambahan)):
    input_data["Tahun"][i] = DataPertambahan['Tahun'].iloc[i]

jumlah = 0
for j in range(len(DataPertambahan)):
    jumlah += DataPertambahan['Pertambahan'].iloc[j]
    input_data["Jumlah"][j] = jumlah
JumlahTiapTahun = pd.DataFrame(data=input_data)


DataPertambahan.plot(kind="line",x="Tahun",y="Pertambahan",title="Pertambahan Game Tiap Tahun")
JumlahTiapTahun.plot(kind="line",x="Tahun",y="Jumlah",title="Jumlah Game Tiap Tahun")
```

#### Menampilkan Data Kesediaan Platform

```python
#pertambahan untuk tiap konsol

# LOADER

sorted_console = sample.value_counts("Platform")

input_data_konsol = {}
input_data_konsol["Jenis Konsol"] = ["a" for i in range(len(sorted_console))]

for j in range(len(sorted_console)):
    input_data_konsol["Jenis Konsol"][j] = sorted_console.index[j]
list_console = pd.DataFrame(data=input_data_konsol)

list_console
```

#### Menampilkan Data Game Untuk Platform Spesifik

```python
pilihan_konsol =int(input("Masukkan pilihan konsol : "))
keyword = list_console['Jenis Konsol'].iloc[pilihan_konsol]

#memuat data konsol
hasil_search = data.loc[data['Platform'] == keyword]
publish_konsol = hasil_search.value_counts('Year')

input_data_konsol = {}
input_data_konsol["Tahun"] = [0 for i in range(len(publish_konsol))]
input_data_konsol["Pertambahan"] = [0 for i in range(len(publish_konsol))]

for i in range(len(publish_konsol)):
    input_data_konsol["Tahun"][i] = publish_konsol.index[i]
for j in range(len(publish_konsol)):
    input_data_konsol["Pertambahan"][j] = publish_konsol.iloc[j]
data_konsol = pd.DataFrame(data=input_data_konsol)

data_konsol_sorted = data_konsol.sort_values('Tahun',ascending=True)

input_data = {}
input_data["Tahun"] = [0 for i in range(len(data_konsol_sorted))]
input_data["Jumlah"] = [0 for i in range(len(data_konsol_sorted))]

for i in range(len(data_konsol_sorted)):
    input_data["Tahun"][i] = data_konsol_sorted['Tahun'].iloc[i]
jumlah = 0

for j in range(len(data_konsol_sorted)):
    jumlah += data_konsol_sorted['Pertambahan'].iloc[j]
    input_data["Jumlah"][j] = jumlah
JumlahTiapTahun = pd.DataFrame(data=input_data)

title1 = "Produksi Game Platform " + keyword
title2 = "Jumlah Game Platform "+ keyword
data_konsol.sort_values('Tahun').plot(kind="line",x="Tahun",y="Pertambahan",title=title1)
JumlahTiapTahun.plot(kind="line",x="Tahun",y="Jumlah", title=title2)
```

#### Menampilkan Peningkatan Jumlah Game di Semua Platform

```python
# Menunjukkan semua perkembangan game untuk semua konsol

for i in range(len(list_console)):
    keyword = list_console['Jenis Konsol'].iloc[i]

    #memuat data konsol
    hasil_search = data.loc[data['Platform'] == keyword]
    publish_konsol = hasil_search.value_counts('Year')

    input_data_konsol = {}
    input_data_konsol["Tahun"] = [0 for i in range(len(publish_konsol))]
    input_data_konsol["Pertambahan"] = [0 for i in range(len(publish_konsol))]

    for i in range(len(publish_konsol)):
        input_data_konsol["Tahun"][i] = publish_konsol.index[i]
    for j in range(len(publish_konsol)):
        input_data_konsol["Pertambahan"][j] = publish_konsol.iloc[j]
    data_konsol = pd.DataFrame(data=input_data_konsol)

    data_konsol_sorted = data_konsol.sort_values('Tahun',ascending=True)

    input_data = {}
    input_data["Tahun"] = [0 for i in range(len(data_konsol_sorted))]
    input_data["Jumlah"] = [0 for i in range(len(data_konsol_sorted))]

    for i in range(len(data_konsol_sorted)):
        input_data["Tahun"][i] = data_konsol_sorted['Tahun'].iloc[i]
    jumlah = 0

    for j in range(len(data_konsol_sorted)):
        jumlah += data_konsol_sorted['Pertambahan'].iloc[j]
        input_data["Jumlah"][j] = jumlah
    JumlahTiapTahun = pd.DataFrame(data=input_data)

    title1 = "Produksi Game Platform " + keyword
    title2 = "Jumlah Game Platform "+ keyword
    data_konsol.sort_values('Tahun').plot(kind="line",x="Tahun",y="Pertambahan",title=title1)
    JumlahTiapTahun.plot(kind="line",x="Tahun",y="Jumlah", title=title2)
```
